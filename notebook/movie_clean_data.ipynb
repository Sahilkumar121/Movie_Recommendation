## Pandas import

```python
import  pandas as pd
```

## Load movie and credit data csv

```python
movie_data = pd.read_csv("../data/movie.csv")
credit_data = pd.read_csv("../data/credit.csv")
```

```python
combine_data = pd.merge(movie_data,credit_data, on="title")
```

```python
combine_data.head(3)
```

### Columns to drop:
### budget, homepage, popularity, release_date, revenue, runtime, spoken_language, original_language, status, vote_average, vote_count, id, movie_id, production_company, production_countries

```python
combine_data = combine_data.drop(["budget", "homepage", "popularity", "release_date", "revenue", "runtime", "spoken_languages", "original_language", "status", "vote_average", "vote_count", "original_title", "keywords", "movie_id", "production_companies", "production_countries", "tagline"], axis=1)
```

```python
combine_data = combine_data.dropna(subset=["overview"])
```

```python
combine_data.isnull().sum()
```

```python
import json
```

```python
def pick_name_genre(obj: str) -> list:
    actual_obj_list = json.loads(obj)

    name_list = []
    for dic in actual_obj_list:
        name_list.append(dic["name"])
    return name_list

def pick_name_cast(obj: str) -> list:
    actual_obj_list = json.loads(obj)

    name_list = []
    count = 0
    for dic in actual_obj_list:
        if count < 3:
            name_list.append(dic["name"])
            count += 1
        else:
            break

    return name_list

def pick_director_crew(obj: str) -> list:
    actual_obj_list = json.loads(obj)

    director_name = []
    for dic in actual_obj_list:
        if dic["job"] == "Director":
            director_name.append(dic["name"])
            break
    return director_name
```

```python
combine_data["genres"] = combine_data["genres"].apply(pick_name_genre)
```

```python
combine_data["cast"] = combine_data["cast"].apply(pick_name_cast)
```

```python
combine_data["crew"] = combine_data["crew"].apply(pick_director_crew)
```

```python
combine_data["overview"] = combine_data["overview"].apply(lambda x : x.split())
```

```python
combine_data["tagline"] = combine_data["genres"] + combine_data["overview"] + combine_data["cast"] + combine_data["crew"]
```

```python
combine_data["tagline"] = combine_data["tagline"].apply(lambda x : [word.lower() for word in x])
```

```python
combine_data.sample(5)
```

```python
final_data = combine_data.drop(["genres", "overview", "crew", "cast"], axis=1)
```

```python
final_data["tagline"] = final_data["tagline"].apply(lambda x: " ".join(x))
```
